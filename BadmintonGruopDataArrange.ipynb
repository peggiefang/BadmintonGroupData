```python
def removeChar(s, delChars):
    for c in delChars:
        s = s.replace(c,'')
    return s
```

```python
colnames=['Date', 'name', 'name2', 'name3', 'name4', 'name5', 'name6', 'name7', 'name8', 'name9', 'name10', 'name11', 'name12', 'name13', 'name14', 'name15', 'name16'] 

att_data = pd.read_csv('/Users/chiehfang/Desktop/badmintonattend.csv',names=colnames)
# Show data
att_data
```

```python
att_datac = pd.read_csv('/Users/chiehfang/Desktop/badmintonattend.csv',header=None,index_col=0)

att_datac
```

```python
attdic = dict()
from datetime import datetime 

for member in members:
    attdic[member]=[]    
    for i in range(1,17):
        day = att_datac.index[att_datac[i] == member].tolist()
        attdic[member] = attdic[member]+day
        attdic[member].sort(key=lambda date: datetime.strptime(date, "%m/%d"))

print(attdic)
```

```python
d1 = {}

for c in att_data.columns[1:]:
    a = att_data[c].value_counts()
    d2 = dict(a)
    for k,v in d2.items():
        if k in d1.keys():
            d1[k] += v
        else:
            d1[k] = v

d1 = pd.DataFrame(list(d1.items()),columns = ['name','count'])
d1['group']=np.nan
# d1.shape[0]
for i in range(d1.shape[0]):
    if d1.loc[i]['name'] in members:
        d1.loc[i,'group'] = 'member'
    else:
        d1.loc[i,'group'] = 'PT'
d1 = d1.sort_values(by=['group'])
d1
# print(d1)

# loc[i,'group'] 用法： https://blog.csdn.net/qq_33711966/article/details/79902276

fliter = d1['group']=='member'
d1[fliter]
```
